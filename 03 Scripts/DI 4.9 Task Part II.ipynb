# 04. Answers

```python
# Histogram Order_hour_of-day
hist = df_full['hour_of_order'].plot.hist(bins = 24)
```

The above histogram shows the frequency of each hour that an order occurs. From this, it is clear that the most orders happen between hours 10 and 15, gradually declining as the evening progresses. Most orders occur during the average person's work day.

```python
# Bar Chart Loyalty Flag
bar_lofg = df_full['loyalty_flag'].value_counts().sort_index().plot.bar()
```

```python
# line graph prices/hour_of_Order
## Random sample generator
np.random.seed(6)
dev2 = np.random.rand(len(df_full)) <= 0.7 #Using standard 70/30 split

big2 = df_full[dev2]
small2 = df_full[~dev2]

df_2 = small2[['hour_of_order','prices']]
```

```python
# Plotting Line
line2 = sns.lineplot(data = df_2, x = 'hour_of_order',y = 'prices')
```

The Graph shows that the price of items ordered is at its lowest during hour 10, but is at its highest early in the morning.

```python
#line graph age/family 
#using same generator as last graph
df_2 = small2[['Age','dependants']]
```

```python
line3 = sns.lineplot(data = df_2, x = 'Age', y = 'dependants')
```

The graph does not show a visible link between age and the number of dependants.

```python
# scatter plot age/income
scat2 = sns.scatterplot(x = 'income', y = 'Age', data = df_full)
```

The plot shows that the older you are the more likely you are to have higher income, but most InstaCart customers have an income between 50,000 and and 200,000.
