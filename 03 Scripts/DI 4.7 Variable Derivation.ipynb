## 02. Creating User Defined Functions

```python
# Limiting Data Frame to Conserve Memory
df_mil = df_full[:1000000]
df_mil.head()
```

```python
# Creating Function
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Applying Function to Data Frame
df_mil['price_range'] = df_mil.apply(price_label, axis = 1)
```

```python
df_mil.head()
```

```python
# creating the same function with loc
df_mil.loc[df_mil['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_mil.loc[(df_mil['prices'] <= 15) & (df_mil['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
 df_mil.loc[df_mil['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# 02. 
result = []
for value in df_mil['orders_day_of_week']:
    if (value == 0) or (value == 1):
        result.append('Busiest Days')
    elif (value == 3) or (value == 4):
        result.append('Slowest Days')
    else:
        result.append('Average Days')
```

compared to the first itteration, this one has a more even spread across categories. This allows us to tell that there are more busy days than slow days during the week.

```python
# 04. early morning = 0-7, mid-day = 8-15, evening to night = 16-23

def hour_grouping(row):

    if row['hour_of_order'] <= 7:
        return 'early morning'
    elif (row['hour_of_order'] >= 8) and (row['hour_of_order'] <= 15):
        return 'mid-day'
    elif row['hour_of_order'] >= 16:
        return 'evening to night'
    else:
        return 'Missing'
```

```python
df_mil['grouping'] = df_mil.apply(hour_grouping, axis = 1)
```

```python
df_mil.head()
```

```python
# mid-day is busiest period
df_mil['grouping'].value_counts()
```

```python
result = []
for value in df_mil['grouping']:
    if value == ('mid-day'):
        result.append('Busiest Period')
    elif value == ('early morning') :
        result.append('Slowest Period')
    else:
        result.append('Average Period')
```

```python
result
```

```python
df_mil['busiest_period'] = result
```

```python
df_mil['busiest_period'].value_counts()
```
