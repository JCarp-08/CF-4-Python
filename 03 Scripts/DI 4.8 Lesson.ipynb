# 04. transform()

```python
full_merge['max_order'] = full_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
full_merge.head()
```

# 05. loyalty flag

```python
full_merge.loc[full_merge['max_order'] > 40 , 'loyalty_flag'] = 'loyal customer'
```

```python
full_merge.loc[(full_merge['max_order'] <= 40) & (full_merge['max_order'] > 10), 'loyalty_flag'] = 'regular customer'
```

```python
full_merge.loc[full_merge['max_order'] <= 10, 'loyalty_flag'] = 'new customer'
```
