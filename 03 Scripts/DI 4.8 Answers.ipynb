```python
# Step 2 & 3
df_full.groupby('department_id').agg({'order_number': ['mean', 'max', 'min']})
```

```python
# Step 5
df_full.groupby('loyalty_flag').agg({'prices': ['mean', 'max', 'min']})
```

mean is lower for loyal customers

```python
# Step 6
# creating mean price flag
df_full['mean_price'] = df_full.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
df_full.head()
```

```python
df_full.loc[df_full['mean_price'] < 10, 'spending_flag'] = 'low spender'
df_full.loc[df_full['mean_price'] >= 10, 'spending_flag'] = 'high spender'
```

```python
df_full.tail()
```

side note: attempted to clean NaNs ran into memory error will try again later

```python
# Step 7
# creating flag column
df_full['med_dspo'] = df_full.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
df_full.head()
```

```python
df_full.loc[df_full['med_dspo'] > 20, 'frequent_shopper'] = 'non-frequent'
df_full.loc[(df_full['med_dspo'] > 10) & (df_full['med_dspo'] <= 20), 'frequent_shopper'] = 'regular'
df_full.loc[df_full['med_dspo'] <= 10, 'frequent_shopper'] = 'frequent'
```
